```python
import siibra
import json
from siibra.core.json_encoder import JSONEncoder
atlas=siibra.atlases['human']
p=atlas.parcellations[0]
```

## Flat JSON encoding

```python
with JSONEncoder() as handle:
    r = p.decode_region('hoc1 left')
    output = handle.get_json(r, detail=True, space=siibra.spaces['mni 152'])
    assert output == {
        "payload": {
            "name": "Area hOc1 (V1, 17, CalcS) left",
            "children": [],
            "rgb": [
                190,
                132,
                147
            ],
            "@id": "minds/core/parcellationregion/v1.0.0/5151ab8f-d8cb-4e67-a449-afe2a41fb007",
            "@type": "minds/core/parcellationregion/v1.0.0",
            "centroids": [{
                "@id": "18a2fe72-e430-447e-8c4d-a08012ea02d4" # @id auto generated if one does not exist
            }]
        },
        "references": [{
            "@id": "18a2fe72-e430-447e-8c4d-a08012ea02d4",
            "@type": "https://openminds.ebrains.eu/sands/CoordinatePoint",
            "coordinateSpace": {}, # even though coordinate space is defined, depth == 1 meant that json encoder will not encode any deeper
            "coordinates": [
                {
                    "@id": None,
                    "@type": "https://openminds.ebrains.eu/core/QuantitativeValue",
                    "value": -8.515510862347028,
                    "unit": {
                        "@id": "id.link/mm"
                    }
                },
                    {
                    "@id": None,
                    "@type": "https://openminds.ebrains.eu/core/QuantitativeValue",
                    "value": -82.38838819846315,
                    "unit": {
                        "@id": "id.link/mm"
                    }
                },
                {
                    "@id": None,
                    "@type": "https://openminds.ebrains.eu/core/QuantitativeValue",
                    "value": 2.00540745659805,
                    "unit": {
                        "@id": "id.link/mm"
                    }
                }
            ]
        }]
    }
```
